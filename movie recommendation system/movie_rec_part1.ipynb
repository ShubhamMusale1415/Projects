```python
# creating convert function for converting list of objects into the list of string 
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
        
        
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if(counter!=3):
            L.append(i["name"])
            counter+=1
        else:
            break
    return L
        
```

```python
def fetchdirector(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if(i["job"]=="Director"):
            L.append(i["name"])
            break;
    return L
        
        
```

```python
from nltk.stem.porter import PorterStemmer 
```

```python
ps=PorterStemmer()
```

```python
def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
        
    
```

```python
df["tags"]=df["tags"].apply(stem)
```

```python
df.head()
```

```python
cv=CountVectorizer(stop_words=None,max_features=5000)
```

```python
xm=cv.fit_transform(df["tags"]).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
cx=cosine_similarity(xm)
```

```python
sorted(list(enumerate(cx[0])),reverse=True,key=lambda x:x[1])[1:6]
# What is x?
# The x is a parameter in the lambda function (lambda x: x[1]).

# Breaking it down:
# enumerate(cx[0]) generates tuples of (index, value) from cx[0].

# Example: If cx[0] = [10, 30, 20], then:

# python
# list(enumerate(cx[0])) → [(0, 10), (1, 30), (2, 20)]
# lambda x: x[1]

# Here, x refers to each tuple in the list, e.g., (0, 10), (1, 30), etc.

# x[1] extracts the second element (the value) for sorting.

# For (0, 10), x[1] = 10

# For (1, 30), x[1] = 30

# What does key=lambda x: x[1] do?
# It tells Python to sort the tuples based on their values (not indices).

# Without it, Python would sort by the first element (index) instead.

# Final Output (Example)
# python
# # Input: cx[0] = [10, 30, 20]
# # After sorting (descending order by value):
# [(1, 30), (2, 20), (0, 10)]
# Summary:
# x is each (index, value) tuple in the list.

# x[1] picks the value (e.g., 30 in (1, 30)).

# The sorted() function uses these values to determine the order.
```

```python
# Pickle in Python - Simple Explanation
# Pickle is a built-in Python library used to:

# Serialize (convert Python objects like lists, dictionaries, or even ML models into a byte stream).

# Deserialize (convert the byte stream back into the original Python object).

# What Does "Pickle" Do?
# Think of it like a food preservative:

# Pickling (Saving): You "preserve" a Python object (e.g., a trained AI model) into a file.

# Unpickling (Loading): Later, you "rehydrate" that file back into a usable Python object.

# Example: Saving & Loading a Python Object
# 1. Pickling (Saving to a File)
# python
# import pickle  

# data = {"name": "Alice", "age": 25, "city": "New York"}  

# # Save to a file
# with open("data.pkl", "wb") as file:  
#     pickle.dump(data, file)  # "wb" = write in binary mode
# This creates a file data.pkl containing the byte version of data.

# 2. Unpickling (Loading from File)
# python
# import pickle  

# # Load back the data
# with open("data.pkl", "rb") as file:  # "rb" = read binary
#     loaded_data = pickle.load(file)  

# print(loaded_data)  # Output: {'name': 'Alice', 'age': 25, 'city': 'New York'}
# You get back the exact same dictionary!

# When is Pickle Used?
# ✅ Saving ML models (Scikit-learn, PyTorch models).
# ✅ Storing complex Python objects (e.g., game progress).
# ✅ Fast data caching (instead of re-processing).
```
